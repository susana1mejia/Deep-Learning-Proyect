## Predicción de ojos abiertos o cerrados a partir de señales electroencefalográficas ##




Los datos utilizados provienen de un dataset de [kaggle](https://https://www.kaggle.com/datasets/robikscube/eye-state-classification-eeg-dataset)





### Características de los datos ###

1.   Catorce señales electroencefalográficas fueron recolectadas de un voluntario usando la configuración de la figura 1 usando electrodos de superficie.
2.  La duración del ensayo fue de 117 segundos
3.   De acuerdo con el registro y cámaras de vídeo, los datos fueron etiquetados como:


> 0 = Ojos abiertos

> 1 = Ojos cerrados

Figura 1. Disposición de electrodos durante el ensayo.

## Objetivo de modelado ##

Usar redes neuronales recurrentes para predecir, en los instantes siguientes de las señales (recuadro verde de la figura 2), la probabilidad de tener los ojos abiertos o cerrados.
